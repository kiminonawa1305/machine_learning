#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
size = 30
label = 'stars'
marker='*'

age = df.loc[:,'age']
rating = df.loc[:, 'refletter.rating']

plt.scatter(age, rating, s=size, label=label, marker=marker, color='red')
plt.xlabel('age')
plt.ylabel('refletter.rating')
plt.title('represent the relationship between age and refletter.rating')
plt.legend()
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
